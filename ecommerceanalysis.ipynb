# Q How many orders per month?

```python
df.groupby('Month').count()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Month')
plt.ylabel('Number of orders')
plt.title('Number of orders for different months')
df.Month.value_counts().plot(kind='bar')
plt.show()
```

```python
df.groupby('Day').count()
```

```python
plt.xlabel('Day')
plt.ylabel('Number of orders')
plt.title('Number of orders for different days')
df.Day.value_counts().plot(kind='bar',cmap='spring')
plt.show()
```

```python
df.groupby('Hour').count()
```

```python
plt.xlabel('Hour')
plt.ylabel('Number of orders')
plt.title('Number of orders for different hours')
df.Hour.value_counts().plot(kind='bar',cmap='autumn')
plt.show()
```
